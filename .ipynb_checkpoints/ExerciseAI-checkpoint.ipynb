```python
# All packages used
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data from csv file
df = pd.read_csv('C:\\Users\\Admin\\Documents\\Chương trình-TTNT\\Chương trình\\spam_Mail_Ai-main\\models\\spam.csv', usecols=[0, 1], encoding='latin-1')
print(df.head())
```

```python
# Rename columns
df.rename(columns={'v1': 'Category', 'v2': 'Message'}, inplace=True)

df['type'] = df.apply(lambda row: 1 if row.Category == 'ham' else 0, axis=1)
print(df.head())
```

```python
#data separation

# list X includes message in database
# list Y includes type in database
print("\n X : \n")
X = list(df['Message'])
print(X)
print("\n Y : \n")
y = list(df['type'])
print(y)
```

```python
# # Split data to 80% data_train and 20% data_test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
# define algorithm
knn = KNeighborsClassifier(n_neighbors=10)

# Instantiate CountVectorizer
vect = CountVectorizer()

# Training vocabulary 
# filter all string that appear in the data and sort it
vect.fit(X_train)

print(vect.vocabulary_)
```

```python
# Term frequancy
# Check the frequancy of strings in a message
X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)

print("X_train_dtm:")
print(X_train_dtm.toarray())
print("\nX_test_dtm:")
print(X_test_dtm.toarray())

# nhiều chuỗi quá nên không rõ ràng :v
```

```python
# Learning KNN by X_train

knn.fit(X_train_dtm, y_train)
```

```python
# Prediction by X Test

y_pred = knn.predict(X_test_dtm)
```

```python
# Compare the prediction result with its target
scoreAfterTrainning = accuracy_score(y_pred, y_test)
print(f"Accuracy: {scoreAfterTrainning * 100:.2f}%")
```

```python
# Message to test
# new = ["This message is brought to you by GMW Ltd. and is not connected to the"]
new = ["URGENT! Your Mobile number has been awarded with a å£2000 prize GUARANTEED"]
dtm = vect.transform(new)
typeofMessage = knn.predict(dtm.toarray())
print(dtm)
```

```python
# Show result

print(f"Accuracy: {scoreAfterTrainning * 100:.2f}%")
if typeofMessage == 0:
    print("\""+new[0]+"\" This message is spam")
else:
    print("\""+new[0]+"\" This message is not spam")
```
